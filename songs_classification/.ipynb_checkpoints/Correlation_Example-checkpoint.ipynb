```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [2, 3, 4, 5, 6],
    'C': [5, 4, 3, 2, 1]
}
df = pd.DataFrame(data)

# Compute the correlation matrix
corr_matrix = df.corr()

# Display the correlation matrix
print(corr_matrix)

# Visualize the correlation matrix using a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Scatter plot of the variables
plt.scatter(df['A'], df['C'], color='blue', label='A vs C')
plt.scatter(df['B'], df['C'], color='red', label='B vs C')

# Adding labels and title
plt.xlabel('Variable A & B')
plt.ylabel('Variable C')
plt.title('Scatter Plot of Variables')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Display first 5 rows of the dataset
print(iris.head())

# Create pairplot of the Iris dataset
sns.pairplot(iris, hue='species', markers=['o', 's', 'D'])
plt.suptitle('Pairplot of the Iris Dataset', y=1.02)
plt.show()
```

```python
# Scatter plot with regression line
sns.regplot(x='sepal_length', y='petal_length', data=iris, color='blue', line_kws={'color':'red'})

# Adding labels and title
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Regression Plot: Sepal Length vs Petal Length')
plt.grid(True)
plt.show()
```

```python
import statsmodels.api as sm

# Defining the dependent and independent variables
X = iris['sepal_length']  # Independent variable
y = iris['petal_length']   # Dependent variable

# Adding a constant to the model (intercept)
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Get the summary of the regression
model_summary = model.summary()
print(model_summary)
```
